Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Jitesh Jangam\\Desktop\\Folder Hub\\Ai and Data Science\\Project\\Data Analysis\\Project\\Netflix User Behavior Analytics\\Raw_Dataset\\reviews.csv")
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip()
```

```python
# STEP 2: Remove duplicate rows
# -------------------------------
df = df.drop_duplicates()
# -------------------------------
# STEP 4: Clean text columns
# (remove spaces inside text)
# -------------------------------
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()

# -------------------------------
# STEP 5: Handle missing values
# numeric → median
# text → "Unknown"
# -------------------------------
for col in df.columns:
    if df[col].dtype != "object":    # numeric
        df[col] = df[col].fillna(df[col].median())
    else:                            # text
        df[col] = df[col].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df['is_verified_watch'] = df['is_verified_watch'].astype(int)
df.head()
```

```python
# -------------------------------
# STEP 6: Save cleaned CSV
# -------------------------------
df.to_csv("reviews_cleaned.csv", index=False)

print("Cleaning completed! File saved as reviews_cleaned.csv")
```
