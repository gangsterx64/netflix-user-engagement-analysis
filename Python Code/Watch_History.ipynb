Data Cleaning

```python
import pandas as pd

# -----------------------------
# Load CSV
# -----------------------------
df = pd.read_csv("C:/Users/Jitesh Jangam/Desktop/Folder Hub/Ai and Data Science/Data Analysis/Project/Netflix User Behavior Analytics/Raw_Dataset/watch_history.csv")

# Remove duplicates
df = df.drop_duplicates()

# Clean column names
df.columns = df.columns.str.strip()

# Clean text columns
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()

# Handle missing values
for col in df.columns:
    if df[col].dtype != "object":
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna("Unknown")

# Save cleaned file
df.to_csv("watch_history_cleaned.csv", index=False)

print("watch_history cleaned successfully!")
```
