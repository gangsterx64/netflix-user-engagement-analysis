Data Cleaning

```python
import pandas as pd

# -------------------------------
# STEP 1: Load the CSV file
# -------------------------------
df = pd.read_csv("C:/Users/Jitesh Jangam/Desktop/Folder Hub/Ai and Data Science/Data Analysis/Project/Netflix User Behavior Analytics/Raw_Dataset/recommendation_logs.csv")   # use your file name

# -------------------------------
# STEP 2: Remove duplicate rows
# -------------------------------
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
# -------------------------------
# STEP 3: Clean column names
# (remove spaces before/after names)
# -------------------------------
df.columns = df.columns.str.strip()
```

```python
# -------------------------------
# STEP 4: Clean text columns
# (remove spaces inside each value)
# -------------------------------
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()

# -------------------------------
# STEP 5: Fill missing values
# numeric → median
# text → "Unknown"
# -------------------------------
for col in df.columns:
    if df[col].dtype != "object":   # numeric column
        df[col] = df[col].fillna(df[col].median())
    else:                           # string column
        df[col] = df[col].fillna("Unknown")
```

```python
# -------------------------------
# STEP 6: Save cleaned CSV
# -------------------------------
df.to_csv("recommendation_logs_cleaned.csv", index=False)

print("CSV file cleaned and saved as recommendation_logs_cleaned.csv")
```
