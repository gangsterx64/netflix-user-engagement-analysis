```python
import pandas as pd

# -----------------------------
# Load CSV
# -----------------------------
df = pd.read_csv("C:\\Users\\Jitesh Jangam\\Desktop\\Folder Hub\\Ai and Data Science\\Project\\Data Analysis\\Project\\Netflix User Behavior Analytics\\Raw_Dataset\\search_logs.csv")

# Remove duplicates
df = df.drop_duplicates()

# Clean column names
df.columns = df.columns.str.strip()

# Clean text columns
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()

df['had_typo'] = df['had_typo'].astype(int)
df['used_filters'] = df['used_filters'].astype(int)

# Handle missing values
for col in df.columns:
    if df[col].dtype != "object":
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna("Unknown")

# Save cleaned file
df.to_csv("search_logs_cleaned.csv", index=False)

print("search_logs cleaned successfully!")
```
